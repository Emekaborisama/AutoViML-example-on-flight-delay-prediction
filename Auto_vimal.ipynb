```
import pandas as pd
import numpy as np
from autoviml.Auto_ViML import Auto_ViML
```

```
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')

#load submission file
sub = pd.read_csv('SampleSubmission.csv')
#read dataset
```

## Requires no data prep - 
you can send in your entire dataframe without a single change. It takes in dirty data!

```
target = 'target'
sample_submission= sub
scoring_parameter = 'balanced-accuracy'
```

```
m, feats, trainm, testm = Auto_ViML(train, target, test, sample_submission,
                                    scoring_parameter = '',
                                    hyper_param='GS',feature_reduction=False,
                                     Boosting_Flag='CatBoost', KMeans_Featurizer = False,Binning_Flag=False,
                                    Add_Poly=0, Stacking_Flag=False,                                    
                                    Imbalanced_Flag=False,
                                    verbose=2)
```

## Requires no categorization of features. 

Other automl library usually request you identify  numeric, categorical, NLP text, date variables etc. 

autoVimal detects all of these automatically as any AutoML package should!

## Does feature reduction automatically:

*Feature reduction param* is set to ture and you can set it to False

 All other packages use all the variables you send them which means they are either bloated or highly complex models. 
 Autovimal is the only package that produces the best model with the fewest variables. That means very easy to interpret!
